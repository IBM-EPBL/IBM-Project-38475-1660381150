## 2. Use .format() to print the following string. 

### Output should be: The diameter of Earth is 12742 kilometers.

```python
planet = "Earth"
diameter = 12742

txt="The diameter of "+planet+" is {ans} kilometers";
print(txt.format(ans=diameter));
```

## 8. Create a dataframe with 3 rows and 2 columns

```python
import pandas as pd
data=[['philp',20],['John',22],['Justin',42]];
print("People name with age");
display=pd.DataFrame(data,columns=["Name","Age"])
print(display)
```

## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
dates=pd.date_range(start="2023-01-01",end="2023-02-10").to_pydatetime().tolist();
dateList=np.array(dates);
print(dateList);
```
